"1" Задано множину s, що складається з цілих чисел. 
Записати ВИРАЗ, значенням якого є множина, що складається з з елементів множини s, що діляться на 13

```python
s = {13, 14, 15, 18, 26, 29, 33, 36, 39}   #Довільна множина
```

```python
a = {el for el in s if el%13 == 0}
a
```

"1" Записати ВИРАЗ, значенням якого є словник, ключами якого є цілі числа від 14 до 2012 (включно),
що діляться на 7, а ключам відповідають їх квадратні корені

```python
a = {x:x**0.5 for x in range(14, 2013) if x%7 == 0}
a
```

"2" Записати ВИРАЗ, значенням якого є словник, ключами якого є цілі числа від 12 до 2000 (включно),
що не діляться на 3, а ключам відповідають їх квадратні корені

```python
a = {x:x**0.5 for x in range(12, 2001) if x%3}
a
```

Нехай змінна s позначає послідовність цілих, по якій можна ітерувати. 
Написати вираз, що перевіряє, чи правда, що всі елементи послідовності більші за задане число num.

```python
num = 4  #Довільне число
s = (10, 5, 6, 7, 9)

all(el > num for el in s)
```
